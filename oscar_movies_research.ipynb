### Корреляция между победами и номинациями

```python
won = oscar_nominated_movies.filter(regex="^(?!Oscar)\w+won$")
won = sorted(won)
```

```python
nominated = oscar_nominated_movies.filter(regex="^(?!Oscar)\w+nominated$")
nominated = sorted(nominated)
```

```python
for won_, nominated_ in zip(won, nominated):
  if won_[:-len("won")] != nominated_[:-len("nominated")]:
    print(won_, nominated_)
```

```python
for won_, nominated_ in zip(won, nominated):
  print(oscar_nominated_movies[won_].corr(oscar_nominated_movies.Oscar_Best_Picture_won), won_)
  print(oscar_nominated_movies[nominated_].corr(oscar_nominated_movies.Oscar_Best_Picture_won), nominated_)
```

```python
oscar_nominated_movies.drop(nominated, axis=1, inplace=True)
```
