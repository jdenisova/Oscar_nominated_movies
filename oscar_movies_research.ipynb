### Корреляция между победами и номинациями

```python
won = oscar_nominated_movies.filter(regex="^(?!Oscar)\w+won$")
won = sorted(won)
```

```python
nominated = oscar_nominated_movies.filter(regex="^(?!Oscar)\w+nominated$")
nominated = sorted(nominated)
```

```python
for won_, nominated_ in zip(won, nominated):
  if won_[:-len("won")] != nominated_[:-len("nominated")]:
    print(won_, nominated_)
```

```python
for won_, nominated_ in zip(won, nominated):
  print(oscar_nominated_movies[won_].corr(oscar_nominated_movies.Oscar_Best_Picture_won), won_)
  print(oscar_nominated_movies[nominated_].corr(oscar_nominated_movies.Oscar_Best_Picture_won), nominated_)
```

# Разделение выборки на train и test

```python
from sklearn.model_selection import train_test_split
```

```python
data_train, data_test = train_test_split(oscar_nominated_movies, test_size=0.5)
```

```python
datasets = [data_train, data_test]
```

# Обучение

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 5
```

```python
knn = KNeighborsClassifier(metric="manhattan", n_neighbors=k)
```

```python
X_train = data_train.drop("Oscar_Best_Picture_won", axis=1)
```

```python
y_train = data_train["Oscar_Best_Picture_won"]
```

```python
knn.fit(X_train, y_train)
```

```python
X_test = data_test.drop("Oscar_Best_Picture_won", axis=1)
```

```python
y_test = data_test["Oscar_Best_Picture_won"]
```

```python
predicted = knn.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test, predicted))
```

```python
metrics.confusion_matrix(y_test, predicted)
```
